```python
# Height and weight correlation with success
# Filter the dataframe to exclude rows without a medal
data1 = data.dropna(subset=['Medal'])
colors = {'Gold': 'red', 'Silver': 'silver', 'Bronze': 'brown', np.nan: 'blue'}
labels = {'Gold': 'Gold Medal', 'Silver': 'Silver Medal', 'Bronze': 'Bronze Medal', np.nan: 'No Medal'}

scatter_plot = plt.scatter(data1['Height'], data1['Weight'], c=data1['Medal'].map(colors))
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=labels[key], markerfacecolor=colors[key]) for key in colors.keys()]
plt.legend(handles=legend_elements)

plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
import seaborn as sns
data_drop=data[['Sex','Age','Height','Weight','Medal']].dropna()
sns.pairplot(data_drop,hue='Medal')
sns.pairplot(data_drop,hue='Sex')
```

```python
newdata=data.dropna()
newdata
```

```python
after2014data=newdata[newdata['Year']>2014]
```

```python
after2014data=after2014data.drop(['ID','Games','NOC','City','Sport','Name'],axis=1)
after2014data
```

```python
print(after2014data.Medal.unique())
print(after2014data.Medal.value_counts())
```

```python
after2014data.Medal.replace(['Bronze','Gold','Silver'],[3,1,2],inplace=True)
```

```python
after2014data.Sex.replace(['F','M'],[1,0],inplace=True)
```

```python
after2014data.Season.replace(['Summer','Winter'],[1,0],inplace=True)
```

```python
onehot_encoded = pd.get_dummies(after2014data['Event'])
onehot_encoded1 = pd.get_dummies(after2014data['Team'])
newdataafter2014 = pd.concat([after2014data, onehot_encoded,onehot_encoded1], axis=1)

newdataafter2014.drop(['Event','Team'], axis=1, inplace=True)
```

```python
x=newdataafter2014.drop('Medal',axis=1)
y=newdataafter2014['Medal']
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

len(x_train),len(y_train),len(x_test),len(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
import sklearn
model=RandomForestClassifier()
model.fit(x_train,y_train)
from sklearn.metrics import classification_report,accuracy_score
y_pred=model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
#Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,model.predict(x_test))
print("confusion matrix",cm)
print('Precision:', sklearn.metrics.precision_score(y_test,model.predict(x_test), average='macro'))
print('Recall:', sklearn.metrics.recall_score(y_test,model.predict(x_test), average='macro'))
print('F1-Score:', sklearn.metrics.f1_score(y_test,model.predict(x_test), average='macro'))
pred_df=pd.DataFrame({"Actual_Value":y_test,"Predicted_Value":y_pred})
print(pred_df)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
